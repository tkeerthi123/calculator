```python
import tkinter as tk

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_equal():
    current = entry.get()
    try:
        result = eval(current)
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_backspace():
    current = entry.get()
    entry.delete(len(current) - 1)

def button_negate():
    current = entry.get()
    if current and current[0] == '-':
        entry.delete(0)
    elif current:
        entry.insert(0, '-')

def button_square():
    current = entry.get()
    try:
        result = eval(current) ** 2
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# Create the main window
root = tk.Tk()
root.title("Calculator")

# Entry widget to display input and results
entry = tk.Entry(root, width=25, borderwidth=5, font=("Arial", 18))
entry.grid(row=0, column=0, columnspan=6, padx=10, pady=10, ipady=10)  # Adjusted columnspan to accommodate larger entry

# Define the buttons
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+'
]

# Create and place the buttons on the grid
row_val = 1
col_val = 0
for button in buttons:
    if button == '=':
        tk.Button(root, text=button, padx=20, pady=20, font=("Arial", 14), bg="#2196F3", fg="white", command=button_equal).grid(row=row_val, column=col_val, padx=5, pady=5)
    else:
        tk.Button(root, text=button, padx=20, pady=20, font=("Arial", 14), bg="#e0e0e0", command=lambda b=button: button_click(b)).grid(row=row_val, column=col_val, padx=5, pady=5)
    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1

# Additional buttons
tk.Button(root, text="C", padx=20, pady=20, font=("Arial", 14), bg="#f44336", fg="white", command=button_clear).grid(row=row_val, column=0, padx=5, pady=5)
tk.Button(root, text="←", padx=20, pady=20, font=("Arial", 14), bg="#e0e0e0", command=button_backspace).grid(row=row_val, column=1, padx=5, pady=5)
tk.Button(root, text="±", padx=20, pady=20, font=("Arial", 14), bg="#e0e0e0", command=button_negate).grid(row=row_val, column=2, padx=5, pady=5)
tk.Button(root, text="x²", padx=20, pady=20, font=("Arial", 14), bg="#e0e0e0", command=button_square).grid(row=row_val, column=3, padx=5, pady=5)

# Start the GUI main loop
root.mainloop()
```
